## SQLite Connection with Python¶

```python
# pip install sqlalchemy

# sqlalchemy python ile sql toollarının(MySQL, MongoDB, vb) bağlantısını kurabilmek için oluşturulmuş bir library
```

```python
from sqlalchemy import create_engine  # sqalchemy kütüphanesinden create_engine modülünü import ettik.
```

```python
temp_db = create_engine("sqlite:///:memory:")  # geçici bir daatabase oluşturmak için hem de sqlite ile bağlantıyı sağlamak için
```

```python
temp_db.table_names() # içeriğine bakmak için
```

The above engine creates a **Dialect object** tailored towards sqlite, as well as a **Pool object** which will establish a DB-API connection at local memory when a connection request is first received. 

**NOTE :** DB-API is an acronym for DataBase Application Programming Interface and a library that lets Python connect to the database server.

```python
tables = pd.read_html('https://en.wikipedia.org/wiki/World_population')
tables
```

```python
len(tables)
```

```python
tables[6]
```

```python
tables[6].to_sql(name="populations", con=temp_db)  # wikipedia dan okuduğumuz tablolardan 6. sını populations ismi ile geçici oluşturduğum db içine göndermek için
```

```python
temp_db.table_names()
```

**How can I open and read a SQL file in python?**

```python
df = pd.read_sql(sql="populations", con=temp_db)  # geçiçi db'e kaydettiğimiz dosyayı okumak için
df
```

**How can I make a SQL query in python?**

```python
# sql dosyası okurken query yazarak okuma yapmak için

pd.read_sql_query(sql="SELECT Country, Population FROM populations", con=temp_db) 
```

```python
import sqlite3  #sqlite üzerinde bir sample_data_db oluşturduktan sonra  sqlite ile bağlantı kurarak bu db'e dosya kaydetme ve okuma
```

```python
with sqlite3.connect('sample_data.db') as cnnt:
    df.to_sql('sample2', cnnt)
```

```python
query = 'SELECT Country, Population FROM sample2'
```

```python
with sqlite3.connect('sample_data.db') as cnnt:
    df_new = pd.read_sql_query(query, cnnt)
```
