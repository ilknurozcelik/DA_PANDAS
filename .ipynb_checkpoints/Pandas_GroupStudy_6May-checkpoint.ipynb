## Combining DataFrames

```python
data_one = {'A': ['A0', 'A1', 'A2', 'A3'],'B': ['B0', 'B1', 'B2', 'B3']}
data_two = {'C': ['C0', 'C1', 'C2', 'C3'], 'D': ['D0', 'D1', 'D2', 'D3']}
one = pd.DataFrame(data_one)
two = pd.DataFrame(data_two)
```

```python
one
```

```python
two
```

```python
pd.concat([one, two], axis=1)
```

```python
pd.concat([one, two], axis=0)
```

```python
pd.concat([one, two], axis=0, ignore_index=True)
```

```python
two.columns = one.columns
```

```python
pd.concat([one, two], axis=0, ignore_index=True)
```

```python
pd.concat([one, two], keys=["One","Two"])
```

```python
pd.concat([one, two], keys=["One","Two"], names=["outer","inner"])
```

```python
data_one = {'A': ['A0', 'A1', 'A2', 'A3'],'B': ['B0', 'B1', 'B2', 'B3']}
data_two = {'C': ['C0', 'C1', 'C2', 'C3'], 'D': ['D0', 'D1', 'D2', 'D3']}
one = pd.DataFrame(data_one)
two = pd.DataFrame(data_two)
three = pd.DataFrame({'D': ['E0', 'E1', 'E2', 'E3'],'H': ['F0', 'F1', 'F2', 'F3']})
three
```

```python
pd.concat([two,three], axis=0)
```

```python
pd.concat([two,three], axis=0, join="inner")  # join=inner olduğunda sadece ortak olan kolonları döndürür.
```

```python
registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bobo','Claire','David']})
logins = pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','Yolanda','Bobo']})
```

```python
registrations
```

```python
logins
```

```python
pd.merge(registrations,logins)  # how default değeri inner old. içim inner join yaptı.
```

```python
pd.merge(registrations,logins, how="inner", on="name")
```

```python
pd.merge(registrations,logins, how="outer", on="name")
```

```python
pd.merge(registrations,logins, how="outer")
```

```python
pd.merge(registrations,logins, how="left")
```

```python
pd.merge(registrations,logins, how="right")
```
